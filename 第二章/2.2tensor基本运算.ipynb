## 矩阵乘法

矩阵乘法有两种形式，第一种是torch.mm,但是只适用于2维矩阵

另一种是torch.matmul,可以用于多维，推荐，此外也可以使用numpy中的@，效果一样


```python
a
```

```python
b = torch.ones(2,2)
b
```

```python
torch.mm(a,b)
```

```python
torch.matmul(a,b)
```

```python
a @ b
```

```python
a = torch.rand(4,784)
x = torch.rand(4,784)
w = torch.rand(512,784)
(x@w.t()).shape
```

# clamp

clamp()内只有一个参数，表示最小值（min），即小于min的都变为min

clamp()内有两个参数，表示（min,max）,即小于min的变为min,大于max变为max

```python
grad = torch.rand(2,3)*15
grad
```

```python
grad.max()
```

```python
grad.median()
```

```python
grad.clamp(10)
```

```python
grad.clamp(5,10)
```

# gather

### torch.gather(input, dim, index, out = None) -> Tensor

```python
prob = torch.randn(4, 10)
```

```python
idx = prob.topk(dim=1, k=3)
idx
```

```python
idx = idx[1]
idx
```

```python
label = torch.arange(10)+101
label
```

```python
torch.gather(label.expand(4,10), dim=1, index=idx.long())
```
