```python
import torch
import numpy as np
import torch.nn.functional as F
```

# 感知机

## 单层感知机

```python
x = torch.randn(1,10)
w = torch.randn(1,10,requires_grad=True)
x
```

```python
o = torch.sigmoid(x@w.t())
o
```

```python
loss = F.mse_loss(torch.ones(1,1),o)
loss
```

```python
loss.backward()
```

```python
w.grad
```

## 多输出感知机

```python
x = torch.randn(1,10)
w = torch.randn(2,10,requires_grad=True)
```

```python
o = torch.sigmoid(x@w.t())
o
```

```python
loss = F.mse_loss(torch.ones(1,2),o)
loss
```

```python
loss.backward()
```

```python
w.grad
```
