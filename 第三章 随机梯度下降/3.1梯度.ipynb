# MSE

# autograd.grad

```python
x = torch.ones(1)
```

```python
w = torch.full([1],2)
```

```python
mse = F.mse_loss(torch.ones(1), x*w)
mse
```

```python
 torch.autograd.grad(mse,[w])
```

```python
w.requires_grad_()
```

```python
torch.autograd.grad(mse,[w])
```

```python
mse = F.mse_loss(torch.ones(1), x*w)
```

```python
torch.autograd.grad(mse,[w])
```

# loss.backward

```python
mse.backward()
```

```python
w.grad
```

# softmax

```python
a = torch.rand(3)
a
```

```python
a.requires_grad_()
```

```python
p = F.softmax(a,dim=0)
```

```python
p
```

```python
torch.autograd.grad(p[1],[a],retain_graph=True)
```

```python
torch.autograd.grad(p[2],[a])
```
